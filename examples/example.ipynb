# Using the SPECIES-XPS library

The SPECIES-XPS library contains some funtions that are usefull for working with the XPS data in python. At the current stage, this only includes loading data into a DataSet object which holds the Group and Region classes. This notebook show an example of how to use them.

```python
# the package is imported as follows:
import species_xps.core as xps
# We also need some other packages for this demonstration
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DataSet loads a list of files into memory, in this case, a list of onte item.
datalist = ['data.xy']
data = xps.DataSet(datalist)
# By using pring on the DataSet object, a list of the groups and regions contained within will be shown
print(data)
```

```python
#The dataset contain these groups, which in turn contain the regions(in its member attribute).
# To get a shorter notation to the Cu 2p spectrum, we acces it by the following:
Cu2p = data.groups[0].member[0]
#Here we can get the information of this dataset by the printing it
print(Cu2p)
```

```python
# The data is contained within the data attribute, and is in this case multidimensional.
# The x-axis is found in the x attribute. 
plt.plot(Cu2p.x, Cu2p.data.sum(axis=0))
```

```python
# Multidimensional data can generate an offsetaxis
Cu2p.generate_offset_axis()
# Which can be usefull for plotting change over time
plt.plot(Cu2p.offset, Cu2p.data.sum(axis=1))
```
